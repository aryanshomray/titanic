```python
from __future__ import absolute_import, division, print_function, unicode_literals
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#tensorflow
import tensorflow as tf
tf.logging.set_verbosity(tf.logging.ERROR)
```

```python
#Uploading Data
df=pd.read_csv('train.csv')
dt=pd.read_csv('test.csv')
```

```python
#Checking null values
df.isnull().sum()
```

```python
df=df.drop(columns=['Cabin'])
dt=dt.drop(columns=['Cabin'])
```

```python
#Converting male to 1 and female to 0
for i in range(df.shape[0]):
    if df.iat[i,4]=='male':
        df.iat[i,4]=1
    else:df.iat[i,4]=0
        
for i in range(dt.shape[0]):
    if dt.iat[i,3]=='male':
        dt.iat[i,3]=1
    else:dt.iat[i,3]=0
```

```python
#Removing Nan values for Age
for i in range(df.shape[0]):
    if pd.isna(df.iat[i,5]):
        df.iat[i,5]=df[(df.Sex==df.iat[i,4])&(df.Pclass==df.iat[i,2])].Age.mean()
        
for i in range(dt.shape[0]):
    if pd.isna(dt.iat[i,4]):
        dt.iat[i,4]=dt[(dt.Sex==dt.iat[i,3])&(dt.Pclass==dt.iat[i,1])].Age.mean()
```

```python
#Removing Nan in Embarked
for i in range(df.shape[0]):
    if pd.isna(df.iat[i,8]):
        df.iat[i,8]='S'
```

```python
#Dropping Fare
df=df.drop(columns=['Fare'])
dt=dt.drop(columns=['Fare'])
```

```python
df
```

```python
#Converting Embarked into numbers
for i in range(df.shape[0]):
    if df.iat[i,9]=='S':
        df.iat[i,9]=0
    if df.iat[i,9]=='C':
        df.iat[i,9]=1
    if df.iat[i,9]=='Q':
        df.iat[i,9]=2
        
for i in range(dt.shape[0]):
    if dt.iat[i,8]=='S':
        dt.iat[i,8]=0
    if dt.iat[i,8]=='C':
        dt.iat[i,8]=1
    if dt.iat[i,8]=='Q':
        dt.iat[i,8]=2
```

```python
#Dropping unnessary fields
df=df.drop(columns=['Ticket','Name'])
dt=dt.drop(columns=['Ticket','Name'])
```

```python
#Changing Index
df.index=df.PassengerId
df=df.drop(columns=['PassengerId'])
dt.index=dt.PassengerId
dt=dt.drop(columns=['PassengerId'])
```

```python
#Normalizing Age
df.Age=df.Age/100
dt.Age=dt.Age/100
```

```python
l0=tf.keras.layers.Dense(12,input_shape=(6,),activation='relu')
l1=tf.keras.layers.Dense(14,activation='relu')
l2=tf.keras.layers.Dense(12,activation='relu')
l3=tf.keras.layers.Dense(2,activation='sigmoid')
```

```python
model=tf.keras.models.Sequential([l0,l1,l2,l3])
model.compile(optimizer=tf.train.AdamOptimizer(0.01),loss=tf.losses.softmax_cross_entropy,metrics=['mae'])
```
